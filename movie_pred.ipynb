```python
import numpy as np 
import pandas as pd 
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

First step is to join the two data frames, into a single data frame. The data frames will be merged together by the title frame. 

```python
movies = movies.merge(credits, on = 'title')
```

```python
movies.head()
```

```python
movies['original_language'].value_counts()
```

From the data set we will keep the following featuers {genres, id, keywords, title,overview, cast, crew }. We need to create a data frame with three columns, movie id, title, tags. 

Overview, genres, keywords, cast and crew will be joined, to form one column. To three cast will be taken, not all cast, as well as only the director from the crew. 

```python
movies = movies[['movie_id','title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L 
```

```python
movies['genres']= movies['genres'].apply(convert)
```

```python
movies['keywords']= movies['keywords'].apply(convert)
```

```python
movies.head()
```

From the cast we want the name of first three actors 

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
            if counter !=3: 
                L.append(i['name'])
                counter+=1
            else:
                break
    
    return L 
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

For the crew we only want the director. We want the dictionary where the job is Director and we need the name tag from that dictionary. 

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L 
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

The overview now has to be converted into a list, since it is right now a string. 

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

Sam Worthington must be SamWorthington. We do not want to confuse the reccomender system, i.e. Sam can refer Sam Mendes and Sam Worthington. So we must remove the space.

```python
movies['genres']= movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords']= movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast']= movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew']= movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head()
```

Now we need to create a new column 'Tag' and concatenate

```python
movies['tags'] = movies['overview']+ movies['genres'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title', 'tags']]
```

```python
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

Now we need to do text vectorisation. The tags need to be converted into vectors, every movie will becoma a vector, There will be 5000 vectors, when someone likes a movie we will suggest the closest vectors to the movie they select. 

We will use bag of words techniques. All tags will be combined, creating a huge string. In that large string we will get 5000 most common words ( in terms of frequency). 

From the large string w1,w2,w3....w5000. 

Then we will go through each movie m1,m2,m3..m5 and see how many times the most frequent words are present in each movie. 

The result is 5000,5000 space where m1,w1 is a vector. 
Lets say w1 is action, we will then go to the first movie and see if this word is present.

We will not consider stop words (words that help form a sentance, but have no contribution to the meaning) i.e. Are, And, Of 

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

stemming is used to stop repeated words

```python
def stem(text):
    y = [] 

    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
new_df['tags']= new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df['tags'][1]
```

Now The model can be built. 

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

Now each movie is a vector, there are 5000 words so there will be a lot of zeros as the array is sparce. 

```python
cv.get_feature_names()
```

Now we need to calc the distance between the vectors ( movies), the larger the distance the lower the similarity between the vectors. 

We will calculate the Cosine distance, not Euclidean distance. We are trying to find the angle. 
When dealing with data with a large dimensions ( out data set has 5000 dimensional space), Euclidean distance is not as accurate. 

The larger the dimensions the lower the effectivness of euclidean distance. We will use cosine similarity in sckitlearn, to calculate the vector distance between a movie and all the other movies. 

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(similarity[0], reverse = True)
```

We must now code the recommend feature, which gives 5 similar movies for a given movie.

```python
sorted(list(enumerate(similarity[0])), reverse=True, key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']== movie].index[0]
    distances = similarity [movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key = lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    
```

```python
recommend ('Spectre')
```
